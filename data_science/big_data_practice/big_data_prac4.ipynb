```python
import pandas as pd
```

```python
df = pd.read_csv('data/enrolment_3.csv')
#df.loc[(df['status'] == 'not allowed') & (df['room assignment'] != 'not assigned'), :]
df
```

```python
dic_room = dict()
room_list = df.loc[df['room assignment'] != 'not assigned', 'room assignment'].unique()
print(room_list)
for room in room_list:
    if room == 'Auditorium':
        room_name = 'Auditorium-'
    elif room == 'Large room':
        room_name = 'Large-'
    elif room == 'Medium room':
        room_name = 'Medium-'
    elif room == 'Small room':
        room_name = 'Small-'
    
    course_list = list(df.loc[df['room assignment'] == room, 'course name'].unique())
    for idx, course in enumerate(sorted(course_list)):
        dic_room[course] = room_name + str(idx+1)

print(dic_room)
for course in dic_room.keys():
    df.loc[(df['course name'] == course) & (df['status'] != 'not allowed'), 'room assignment'] = dic_room[course]

#df.rename(columns={'room assignment': 'room number'}, inplace=True)
#df.loc[df['status'] == 'not allowed', 'room assignment'].value_counts()
#df.loc[(df['status'] == 'not allowed') & (df['room assignment'] != 'not assigned'), :]
#df['room assignment'].value_counts().index
df
```
